### Задача 1 [1 балл].
Пробег (в 1000км) автомобиля "Лада Калина" до капитального ремонта двигателя является нормальной случайной величиной с неизвестным математическим ожиданием μ и известной дисперсией 49. По выборке из 20 автомобилей постройте доверительный интервал для математического ожидания пробега с уровнем доверия 0.95.

```python
import numpy as np
import scipy.stats as st
```

```python
X = [321.90 , 307.26, 194.18, 373.78, 287.03, 251.33, 249.19, 303.62,
       333.47, 264.23, 277.13, 325.44, 227.72, 299.75, 267.21, 316.72,
       269.45, 260.69, 328.81, 335.34]

sample_mean = np.mean(X)
a, b = st.norm.interval(confidence=0.95, loc=sample_mean, scale=7 / np.sqrt(20))
print(f'Доверительный интервал для математического ожидания: ({np.round(a, 2)}, {np.round(b, 2)})')
print(f'mu = {np.round(sample_mean, 2)} +- {np.round(b - sample_mean, 2)}')
```

### Задача 2 [2 балла].
Бутерброд подбросили 200 раз, 105 раз он упал маслом вниз и 95 маслом вверх. Проверьте гипотезу о том, что бутерброд с равной вероятностью падает маслом вверх и вниз.

Представим выборку в виде нулей и единиц: $X_i = 1$ ⟺ бутерброд упал маслом вниз, $X_i = 0$ ⟺ бутерброд упал маслом вверх. $X_1, ..., X_n$ ~ iid Bern(p). Мы хотим проверить гипотезу $H_0: p = 0.5$ ⟺ бутерброд падает равновероятно.

```python
from statsmodels.stats.proportion import binom_test # бином тест как раз и выполняет проверку того, что вероятность успеха (в эксперименте Бернулли) есть p
### Или любой другой статистический пакет
### Или можете реализовать критерий самостоятельно

p_value = binom_test(105, 200, prop=0.5, alternative='two-sided')
print(f'p_value = {np.round(p_value, 2)}')
```

Таким образом, гипотеза $H_0$ не отвергается при любом уровне значимости $\alpha < p\_value = 0.52$

### Задача 3 [3 балла].
В Законодательном собрании некоторой страны (450 депутатов) обсуждают введение налога на воздух. Агент ФСБ, подслушавший заседание, утверждает, что ЗА планирует проголосовать около 70% депутатов. Закон будет принят, если ЗА проголосует более половины депутатов. Какое минимальное количество депутатов нужно опросить на выходе с заседания, чтобы построить критерий, проверяющий гипотезу о том, что закон не принят, против альтернативы (информации от агента), с уровнем значимости 5% и мощностью не менее 80%?

```python
a = 0.05
b = 1 - 0.8
p0 = 0.5
pa = 0.7

n = ((np.sqrt(pa * (1 - pa)) * st.norm.ppf(1 - b) + np.sqrt(p0 * (1 - p0)) * st.norm.ppf(1 - a)) / (pa - p0)) ** 2
print(f'Искомое минимальное количество депутатов: {int(np.ceil(n))}')
```

Теперь имеем: $e^{-\overline X}$ ~ $N(e^{-λ}, \frac{\hat{λ}}{n} e^{-2\hat{λ}})$ - в дисперсии возьмем состоятельные оценки, ведь асимптотика не поменяется.

$\overline X$ - состоятельная оценка для $MX_1 = λ$. Таким образом, $e^{-\overline X}$ ~ $N(e^{-λ}, \frac{\overline X}{n} e^{-2\overline X})$

```python
sample_mean = np.mean(X)
n = len(X)
a, b = st.norm.interval(confidence=0.95, loc=np.exp(-sample_mean), scale=np.sqrt(sample_mean / n * np.exp(-2 * sample_mean)))
print(f'Доверительный интервал для вероятности, что завтра вечером преподаватель будет трезв, с уровнем доверия 0.95: ({np.round(a, 4)}, {np.round(b, 4)})')
print(f'e^(-lambda) = {np.round(np.exp(-sample_mean), 4)} +- {np.round(b - np.exp(-sample_mean), 4)}')
```
